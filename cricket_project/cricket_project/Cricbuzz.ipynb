```python
#team
import pandas as pd
import json

with open("combined_matches.json") as f:
    all_data = json.load(f)

teams = []

for data in all_data:
    for mtype in data.get('typeMatches', []):
        for s in mtype.get('seriesMatches', []):
            for ms in s.get('seriesAdWrapper', {}).get('matches', []):
                mi = ms.get('matchInfo', {})
                for side in ['team1', 'team2']:
                    t = mi.get(side, {})
                    if t:
                        teams.append((t.get('teamId'), t.get('teamName'), t.get('teamSName', '')))

df_teams = pd.DataFrame(teams, columns=["team_id", "team_name", "short_name"]).drop_duplicates()
df_teams.to_csv("teams.csv", index=False)
df_teams.head()
```

```python
import requests
import mysql.connector

API_KEY = "90a1f9d311msh33b467f3e3ff9ecp1a1aaajsn4597d6b59a9e"

# URLs for different match types
urls = {
    "live": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/live",
    "upcoming": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/upcoming",
    "recent": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
}

headers = {
    "x-rapidapi-key": API_KEY,
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Nithish@12345",
    database="cricbuzz_db"
)
cursor = conn.cursor()

# Create table if not exists
cursor.execute("""
CREATE TABLE IF NOT EXISTS Matches (
    id INT AUTO_INCREMENT PRIMARY KEY,
    match_category VARCHAR(50),
    match_type VARCHAR(50),
    series_name VARCHAR(200),
    format VARCHAR(20),
    match_desc VARCHAR(200),
    team1 VARCHAR(100),
    team1_score VARCHAR(50),
    team2 VARCHAR(100),
    team2_score VARCHAR(50),
    status VARCHAR(200),
    state VARCHAR(50),
    start_date BIGINT,
    end_date BIGINT,
    venue VARCHAR(200),
    city VARCHAR(100)
)
""")
conn.commit()

# Fetch data from API and insert into MySQL
for match_type_key, url in urls.items():
    resp = requests.get(url, headers=headers)
    if resp.status_code != 200:
        print(f"Error fetching {match_type_key} matches: {resp.status_code}")
        continue

    data = resp.json()
    type_matches = data.get('typeMatches', [])

    for match_type_group in type_matches:
        match_type_name = match_type_group.get('matchType', 'N/A')
        series_list = match_type_group.get('seriesMatches', [])

        for series in series_list:
            series_ad_wrapper = series.get('seriesAdWrapper', {})
            series_name = series_ad_wrapper.get('seriesName', 'N/A')
            matches = series_ad_wrapper.get('matches', [])

            for match in matches:
                match_info = match.get('matchInfo', {})
                match_score = match.get('matchScore', {})

                team1 = match_info.get('team1', {}).get('teamName', 'N/A')
                team2 = match_info.get('team2', {}).get('teamName', 'N/A')
                status = match_info.get('status', 'N/A')
                match_desc = match_info.get('matchDesc', 'N/A')
                match_format = match_info.get('matchFormat', 'N/A')
                start_date = match_info.get('startDate', 'N/A')
                end_date = match_info.get('endDate', 'N/A')
                state = match_info.get('state', 'N/A')
                venue = match_info.get('venueInfo', {}).get('ground', 'N/A')
                city = match_info.get('venueInfo', {}).get('city', 'N/A')

                # Scores (if available)
                team1_score_data = match_score.get('team1Score', {}).get('inngs1', {})
                team2_score_data = match_score.get('team2Score', {}).get('inngs1', {})

                team1_score = f"{team1_score_data.get('runs', '')}/{team1_score_data.get('wickets', '')} ({team1_score_data.get('overs', '')} overs)"
                team2_score = f"{team2_score_data.get('runs', '')}/{team2_score_data.get('wickets', '')} ({team2_score_data.get('overs', '')} overs)"

                # Insert into MySQL
                cursor.execute("""
                INSERT INTO Matches (
                    match_category, match_type, series_name, format, match_desc,
                    team1, team1_score, team2, team2_score, status, state,
                    start_date, end_date, venue, city
                ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                """, (
                    match_type_key, match_type_name, series_name, match_format, match_desc,
                    team1, team1_score, team2, team2_score, status, state,
                    start_date, end_date, venue, city
                ))

# Commit changes and close connection
conn.commit()
conn.close()
print("All matches from API inserted into MySQL successfully!")
```

```python
# ==========================================
# VENUE TABLE: Create table and populate from Matches API
# ==========================================

import requests
import mysql.connector

API_KEY = "90a1f9d311msh33b467f3e3ff9ecp1a1aaajsn4597d6b59a9e"

# URLs for Matches
urls = {
    "live": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/live",
    "upcoming": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/upcoming",
    "recent": "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"
}

headers = {
    "x-rapidapi-key": API_KEY,
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Nithish@12345",
    database="cricbuzz_db"
)
cursor = conn.cursor()

# Create VENUE table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Venue (
    venue_id INT AUTO_INCREMENT PRIMARY KEY,
    venue_name VARCHAR(200),
    city VARCHAR(100),
    country VARCHAR(100)
)
""")
conn.commit()

# Extract unique venues from Matches API
venues_set = set()

for match_type_key, url in urls.items():
    resp = requests.get(url, headers=headers)
    if resp.status_code != 200:
        print(f"Error fetching {match_type_key} matches: {resp.status_code}")
        continue

    data = resp.json()
    type_matches = data.get('typeMatches', [])
    for match_type_group in type_matches:
        series_list = match_type_group.get('seriesMatches', [])
        for series in series_list:
            series_ad_wrapper = series.get('seriesAdWrapper', {})
            matches = series_ad_wrapper.get('matches', [])
            for match in matches:
                match_info = match.get('matchInfo', {})
                venue = match_info.get('venueInfo', {})
                venue_name = venue.get('ground', 'N/A')
                city = venue.get('city', 'N/A')
                country = venue.get('country', 'N/A')
                venues_set.add((venue_name, city, country))

# Insert venues into table
for v in venues_set:
    cursor.execute("""
    INSERT IGNORE INTO Venue (venue_name, city, country)
    VALUES (%s, %s, %s)
    """, v)

conn.commit()
print("Venue table populated successfully from Matches data!")

# Close connection
cursor.close()
conn.close()
```
